```python
from sklearn import tree
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree, _tree
from sklearn.model_selection import GridSearchCV
from sklearn.multioutput import MultiOutputClassifier
from skmultilearn.problem_transform import LabelPowerset
from sklearn.naive_bayes import GaussianNB
import chardet
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, train_test_split
import numpy as np
import math
from sklearn.impute import SimpleImputer
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import xgboost as xgb
```

## 1(b) Build a decision tree on the whole data set and plot it.

```python
with open("../data/diagnosis.data", "rb") as f:
    result = chardet.detect(f.read())
encoding = result['encoding']
column_names = [
    "Temperature",
    "Nausea",
    "Lumbar Pain",
    "Urine Pushing",
    "Micturition Pains",
    "Urethra Issues",
    "Bladder Inflammation",
    "Renal Pelvis Nephritis"
]

data = pd.read_csv("../data/diagnosis.data", delimiter=r"\s+", header=None, names=column_names, encoding=encoding)
data["Temperature"] = data["Temperature"].str.replace(",", ".").astype(float)
data = data.replace({"yes": 1, "no": 0})
print(data)


X = data[column_names[:-2]]  
y = data[column_names[-2:]]  


```

```python
tree_clf = MultiOutputClassifier(DecisionTreeClassifier(random_state=0))
tree_clf.fit(X, y)

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

plot_tree(tree_clf.estimators_[0], feature_names=X.columns, filled=True, class_names=['0', '1'], ax=axs[0])
axs[0].set_title("Decision Tree for Bladder Inflammation ")

plot_tree(tree_clf.estimators_[1], feature_names=X.columns, filled=True, class_names=['0', '1'], ax=axs[1])
axs[1].set_title("Decision Tree for Renal Pelvis Nephritis")

plt.tight_layout()
plt.show()
```

```python
lp_classifier = LabelPowerset(classifier=DecisionTreeClassifier(random_state=0), require_dense=[False, False])
lp_classifier.fit(X, y)

fig, axs = plt.subplots(1, 1, figsize=(20, 10))  

plot_tree(lp_classifier.classifier, feature_names=X.columns, filled=True, ax=axs)
axs.set_title("Label Powerset Decision Tree")

plt.tight_layout()
plt.show()
```

# 1(c) Convert the decision rules into a set of IF-THEN rules.

```python
def tree_to_code(tree, feature_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]
    print ("def tree({}):".format(", ".join(feature_names)))

    def recurse(node, depth):
        indent = "  " * depth
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print ("{}if {} <= {}:".format(indent, name, threshold))
            recurse(tree_.children_left[node], depth + 1)
            print ("{}else:  # if {} > {}".format(indent, name, threshold))
            recurse(tree_.children_right[node], depth + 1)
        else:
            print ("{}return {}".format(indent, tree_.value[node]))

    recurse(0, 1)
tree_to_code(lp_classifier.classifier, X.columns)
```

# 1(d) Use cost-complexity pruning to find a minimal decision tree and a set of decision rules with high interpretability.


```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X, y)
plt.figure(figsize=(12, 8))
plot_tree(dt, feature_names=X.columns, class_names=y.columns, filled=True)
plt.title('unpruned tree')
plt.show()
path = dt.cost_complexity_pruning_path(X, y)
ccp_alphas = path.ccp_alphas
impurities = path.impurities
 
plt.figure(figsize=(8, 6))
plt.plot(ccp_alphas, impurities)
plt.xlabel('alpha')
plt.ylabel('impurity')
plt.title('Alpha vs Impurity')
plt.show()
```

```python

param_grid = {'ccp_alpha': ccp_alphas}
scoring = 'accuracy'
gs = GridSearchCV(dt, param_grid, scoring=scoring, cv=5)
gs.fit(X, y)


cv_results = gs.cv_results_



results_df = pd.DataFrame({
    'alpha': cv_results['param_ccp_alpha'],
    'mean_test_score': cv_results['mean_test_score']
})


non_perfect_scores = results_df[results_df['mean_test_score'] < 1.0]


non_perfect_scores = non_perfect_scores.sort_values(by='mean_test_score', ascending=False)

print(non_perfect_scores)
```

```python
alphas = non_perfect_scores['alpha'].values

n_alphas = len(alphas)
n_cols = 2 
n_rows = math.ceil(n_alphas / n_cols)


fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 6, n_rows * 6)) 
axes = axes.flatten()  
for i, alpha in enumerate(alphas):

    dt_pruned = DecisionTreeClassifier(ccp_alpha=alpha, random_state=42)
    dt_pruned.fit(X, y)
    

    if dt_pruned.tree_.node_count > 1:
       
        plot_tree(dt_pruned, feature_names=X.columns, class_names=np.unique(y).astype(str), filled=True, ax=axes[i])
        axes[i].set_title(f'Decision Tree with ccp_alpha={alpha:.4f}')
    else:
     
        axes[i].text(0.5, 0.5, f'Single node tree\nccp_alpha={alpha:.4f}', 
                     horizontalalignment='center', verticalalignment='center')
        axes[i].set_title(f'Decision Tree with ccp_alpha={alpha:.4f}')
        axes[i].axis('off')


for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])


plt.tight_layout()
plt.show()

dt_pruned = DecisionTreeClassifier(ccp_alpha=0.0491, random_state=42)
dt_pruned.fit(X, y)
tree_to_code(dt_pruned, X.columns)
```

## 2(f) Fit a linear model using least squares to the training set and report the test error

```python
linear_model = LinearRegression(fit_intercept=False) 
'''
If the features and target values are scaled or normalized, especially with zero-centered normalization (mean of zero), 
the data might already be positioned close to the origin. In such cases, removing the intercept allows the model to fit 
more closely to the data, as adding an intercept would effectively shift the predictions away from the origin, increasing the error.
'''
linear_model.fit(X_train, y_train)
y_pred = linear_model.predict(X_test)
test_error = mean_squared_error(y_test, y_pred)
print(f"Test Error (Mean Squared Error) without intercept: {test_error:.4f}")
```

## 2(g) Fit a ridge regression model on the training set, with  chosen by cross-validation. Report the test error obtained.


```python
alphas = [0.1, 1.0, 10.0, 50.0, 100.0, 200.0, 500.0, 1000.0]

ridge_cv_model = RidgeCV(alphas=alphas, scoring='neg_mean_squared_error', cv=5)


ridge_cv_model.fit(X_train, y_train)

best_alpha = ridge_cv_model.alpha_
print(f"Best alpha (lambda) chosen by cross-validation: {best_alpha}")


y_pred_ridge = ridge_cv_model.predict(X_test)
test_error_ridge = mean_squared_error(y_test, y_pred_ridge)

print(f"Test Error with Ridge Regression (Mean Squared Error): {test_error_ridge:.4f}")
```

## 2(h) Fit a LASSO model on the training set, with  chosen by cross-validation. Report the test error obtained, along with a list of the variables selected by the model. Repeat with standardized features. Report the test error for both cases and compare them.

```python
lasso_cv_model = LassoCV(cv=5, random_state=42)


lasso_cv_model.fit(X_train, y_train)

y_pred_unscaled = lasso_cv_model.predict(X_test)
test_error_unscaled = mean_squared_error(y_test, y_pred_unscaled)

selected_features_unscaled = X_train.columns[np.abs(lasso_cv_model.coef_) > 1e-5]

print("LASSO without Standardization")
print(f"Best alpha chosen by cross-validation: {lasso_cv_model.alpha_}")
print(f"Test Error (MSE): {test_error_unscaled:.4f}")
print("Selected Variables:", list(selected_features_unscaled))

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

lasso_cv_model_scaled = LassoCV(cv=5, random_state=42)
lasso_cv_model_scaled.fit(X_train_scaled, y_train)


y_pred_scaled = lasso_cv_model_scaled.predict(X_test_scaled)
test_error_scaled = mean_squared_error(y_test, y_pred_scaled)

selected_features_scaled = X_train.columns[np.abs(lasso_cv_model_scaled.coef_) > 1e-5]

print("\nLASSO with Standardization")
print(f"Best alpha chosen by cross-validation: {lasso_cv_model_scaled.alpha_}")
print(f"Test Error (MSE): {test_error_scaled:.4f}")
print("Selected Variables:", list(selected_features_scaled))


print("\nComparison")
print(f"Test Error without Standardization (MSE): {test_error_unscaled:.4f}")
print(f"Test Error with Standardization (MSE): {test_error_scaled:.4f}")
```

## 2(i) Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.

```python
pcr_pipeline = Pipeline([
    ('pca', PCA()),                  #
    ('linear_regression', LinearRegression())  
])


param_grid = {
    'pca__n_components': range(1, X_train.shape[1] + 1)  
}

grid_search = GridSearchCV(pcr_pipeline, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_pcr_model = grid_search.best_estimator_
best_num_components = grid_search.best_params_['pca__n_components']
print(f"Optimal number of principal components (M) chosen by cross-validation: {best_num_components}")

y_pred = best_pcr_model.predict(X_test)
test_error_pcr = mean_squared_error(y_test, y_pred)

print(f"Test Error with PCR (Mean Squared Error): {test_error_pcr:.4f}")
```

## 2(j) In this section, we would like to fit a boosting tree to the data. As in classification trees, one can use any type of regression at each node to build a multivariate regression tree. Because the number of variables is large in this problem, one can use L1-penalized regression at each node. Such a tree is called L1 penalized gradient boosting tree. You can use XGBoost5 to fit the model tree. Determine  (the regularization term) using cross-validation

```python

param_grid = {
    'alpha': [0.01, 0.1, 1, 10, 50, 100]  
}


xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',   
    booster='gbtree',               
    max_depth=3,                    
    learning_rate=0.1,              
    n_estimators=100,               
    reg_alpha=0,                   
    random_state=42                 
)

grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_xgb_model = grid_search.best_estimator_
best_alpha = grid_search.best_params_['alpha']
print(f"Optimal alpha (L1 regularization term) chosen by cross-validation: {best_alpha}")

y_pred = best_xgb_model.predict(X_test)
test_error = mean_squared_error(y_test, y_pred)

print(f"Test Error with L1-Penalized Gradient Boosting Tree (MSE): {test_error:.4f}")
```
